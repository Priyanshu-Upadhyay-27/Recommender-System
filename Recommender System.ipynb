```python
movies = movies[["genres", 'id', 'keywords', 'title', 'overview', 'cast', 'crew']]
```

```python
# Now we will be merging the last 4 or 5 columns to get the tags
movies.iloc[0].genres 
# since it is in a nested list- dictionary format, we have to preprocess it to get the tags 
```

```python
# making a function to preprocess it
import ast
def change1(obj):
    L = [i["name"] for i in ast.literal_eval(obj)] 
    #ast.literal_eval() is used to evaluate a string, i.e. it converts and executes the safe string code, therefore converting the string dictionary into an actual dictionary.
    return L
```

```python
movies["genres"] = movies['genres'].apply(change1)
```

```python
movies["keywords"] = movies['keywords'].apply(change1)
```

```python
movies.head()
```

```python
movies.iloc[0].cast # here each dictionary is representing a cast
```

```python
def change2(obj):
    L = [i['name'] for counter, i in enumerate(ast.literal_eval(obj)) if counter < 4] # choosen top 4 cast
    return L
```

```python
movies["cast"] = movies["cast"].apply(change2)
```

```python
movies.head()
```

```python
movies.iloc[0].crew
```

```python
def change3(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            L.append(i["name"])
            break  # Stop after finding the first match
    return L
```

```python
movies['crew'] = movies['crew'].apply(change3)
```

```python
movies.head()
```

```python
movies['overview'].apply(lambda x : x.split())
```

```python
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
movies.head()
```

```python
movies.shape
```

```python
def remove_spaces(obj):
    L = [i.replace(" ", "") for i in obj]
    return L
```

```python
movies['genres'] = movies['genres'].apply(remove_spaces)
```

```python
movies['keywords'] = movies['keywords'].apply(remove_spaces)
movies['cast'] = movies['cast'].apply(remove_spaces)
movies['crew'] = movies['crew'].apply(remove_spaces)
#spaces are removed to create unique tags, which avoids confusion in column
```

```python
movies.head()
```

```python
movies['tags'] = movies["overview"] + movies["genres"] + movies["cast"] + movies['keywords'] + movies["cast"] + movies['crew']
```

```python
movies.head()
```

```python
# creating a new dataframe
df = movies.drop(columns = ['genres', "cast", "crew", "keywords", "overview"])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray() # we have every movies in vector form, every row is a vector/movie in vectorized form
```

```python
vectors[0]
```

```python
x = cv.get_feature_names_out()
```

```python
print(np.array2string(x, threshold=np.inf))
#used to show full array 
```

```python
# Now We will be using stemming to get faster results, else we can also use lemmetization, but it would be a slower nlp technique
```

```python
from nltk.stem.porter import PorterStemmer
ps =PorterStemmer() # it will  convert any word to its root form
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
df["tags"] = df["tags"].apply(stem)
```

```python
df["tags"]
```

```python
print(np.array2string(cv.get_feature_names_out(), threshold=np.inf)) # this list contains only root words
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
angular_d = cosine_similarity(vectors)
```

```python
angular_d # it is the every angular distance with every other movie in a 5000 dimensional movie plan.
```

```python
sorted(list(enumerate(angular_d[0])), reverse = True, key = lambda x : x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = df[df["title"] == movie].index[0] #retrieving the index of the movie searched
    distances = angular_d[movie_index] # taking that distance of all other movie from that searched movie
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6] # sorting that array with preserved index
    for i in movie_list:
        print(df.iloc[i[0]].title)
```

```python
recommend("The Midnight Meat Train") # working
```
